```python
import pandas as pd
from google.colab import files
uploaded = files.upload()


# Load the dataset
data = pd.read_csv('Orange Quality Data.csv')

# Display dataset information
print(data.info())
print(data.head())
```

Step 2: Correlation Between Size and Weight


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation value
correlation_size_weight = data['Size (cm)'].corr(data['Weight (g)'])
print("Correlation between Size and Weight:", correlation_size_weight)

# Scatterplot
plt.figure(figsize=(8, 6))
sns.regplot(x='Size (cm)', y='Weight (g)', data=data, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'})
plt.title("Size vs Weight")
plt.xlabel("Size (cm)")
plt.ylabel("Weight (g)")
plt.show()
```

Step 3: Sweetness (Brix) vs Acidity (pH)


```python
# Correlation value
correlation_brix_ph = data['Brix (Sweetness)'].corr(data['pH (Acidity)'])
print("Correlation between Sweetness (Brix) and Acidity (pH):", correlation_brix_ph)

# Scatterplot
plt.figure(figsize=(8, 6))
sns.regplot(x='Brix (Sweetness)', y='pH (Acidity)', data=data, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'})
plt.title("Sweetness (Brix) vs Acidity (pH)")
plt.xlabel("Sweetness (Brix)")
plt.ylabel("Acidity (pH)")
plt.show()
```

3) How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?

```python
# Correlation value
correlation_softness_ripeness = data['Softness (1-5)'].corr(data['Ripeness (1-5)'])
print("Correlation between Softness and Ripeness:", correlation_softness_ripeness)

# Scatterplot
plt.figure(figsize=(8, 6))
sns.regplot(x='Softness (1-5)', y='Ripeness (1-5)', data=data, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'})
plt.title("Softness vs Ripeness")
plt.xlabel("Softness (1-5)")
plt.ylabel("Ripeness (1-5)")
plt.show()
```

4) Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?

```python
# Grouping by blemishes and calculating mean values
blemish_analysis = data.groupby('Blemishes (Y/N)')[['Ripeness (1-5)', 'Softness (1-5)']].mean()
print("Impact of Blemishes on Ripeness and Softness:\n", blemish_analysis)

# Visualize using a bar plot
blemish_analysis.plot(kind='bar', figsize=(10, 6))
plt.title("Impact of Blemishes on Ripeness and Softness")
plt.ylabel("Average Levels")
plt.xlabel("Blemishes (Y/N)")
plt.xticks(rotation=0)
plt.show()
```
